```python
features = ['PassengerId','Pclass','Age','SibSp','Parch']
```

```python
dt = DecisionTreeClassifier(min_samples_split=20, random_state=99)
```

```python
y = train_set["Survived"]
X = train_set[features].fillna(method='ffill')
```

```python
dt.fit(X,y)
```
